```python
import xarray as xr
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Read in data
df_v0 = pd.read_csv('v0_series.csv', index_col='time')
df_v1 = pd.read_csv('v1_series.csv', index_col='time')
df_v4 = pd.read_csv('df_v4.csv', index_col='Unnamed: 0')

# Get values from v5
v0_series = df_v0['temp'].values
v1_series = df_v1['temp'].values

# Create global average timeseries for v4
df_v4_annual = df_v4.reindex(sorted(df_v4.columns), axis=1)
v4_series = df_v4_annual.mean(axis=0).values

# Set year ranges
years_v5 = range(1880, 2024)
years_v4 = range(1880, 2025)

# Plot results
years = range(1880, 2024)
plt.plot(years_v5, v0_series, label='v5 (gridding v0)')
plt.plot(years_v5, v1_series, label='v5 (gridding v1)')
plt.plot(years_v4, v4_series, label='v4')
plt.axvline(x=1950)
plt.grid()
plt.legend()
plt.xlabel('Year')
plt.ylabel('Temperature Anomaly (Celsius)')
plt.title('Comparison of v4 vs. v5 gridding methods')
```
